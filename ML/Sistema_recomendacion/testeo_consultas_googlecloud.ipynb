```python
from google.cloud import bigquery
import pandas as pd
import os
```

```python
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "../credencial.json"
```

```python
# Configura tu proyecto y tu dataset
project_id = 'pure-hue-399113'
dataset_id = 'datos_crudos_google'
table_id = 'cualidades_google_crudo'
limite = 10
```

```python
# Inicializa el cliente de BigQuery
client = bigquery.Client()
```

```python
# Construye la consulta SQL
query = f"""
    SELECT *
    FROM `{project_id}.{dataset_id}.{table_id}`
    LIMIT {limite}
"""
```

```python
# Ejecuta la consulta y obt√©n los resultados
query_job = client.query(query)
results = query_job.result()
```

```python
rows = []

# Itera sobre las filas y almacena los datos en la lista
for row in results:
    print(row)
    rows.append(row)
```
