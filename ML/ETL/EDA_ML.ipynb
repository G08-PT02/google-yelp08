```python
def desglose(columna):
    lista_categorias = []
    for row in columna:
        if row:
            palabras = row.split(',')
            for palabra in palabras:
                palabra = palabra.strip()  # Eliminar espacios en blanco alrededor de la palabra
                if palabra not in lista_categorias:
                    lista_categorias.append(palabra)
    return lista_categorias
```
