## Extraccion de palabras clave con KeyBERT

```python
from keybert import KeyBERT

import spacy

model = KeyBERT(model="distilbert-base-nli-mean-tokens")
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('spanish'))
# Agregar palabras adicionales si es necesario
stop_words.update(["los", "las", "un", "una", "unos", "unas"])
```

```python
def extract_keywords(text, num_keywords=10):
    keywords = model.extract_keywords(text, keyphrase_ngram_range=(1, 1), stop_words=None, use_mmr=True, diversity=0.7, top_n=num_keywords)
    return [keyword[0] for keyword in keywords]

user_input = "Me gustan los lugares con comida italiana y buen vino."

# Carga el modelo en espaÃ±ol
nlp = spacy.load("es_core_news_sm")

def lemmatize_text(text):
    doc = nlp(text)
    lemmatized = [token.lemma_ for token in doc]
    return ' '.join(lemmatized)

user_input = "Queremos comer unas buenas pizzas"
lemmatized_input = lemmatize_text(user_input)

print(lemmatized_input)  # DeberÃ­a mostrar "querer plato italiano"

keywords = extract_keywords(lemmatized_input) # Obtenemos las keywords del imput
filtered_keywords = [word for word in keywords if word not in stop_words] #Filtramos para sacar las stop_words

print(filtered_keywords)
```
